```python
import numpy as np
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
x=[1,7,9]
y=[1,5,8]
```

```python
plt.plot(x,y)
```

```python
x=[1,3,4]
y=[2,3,4]
plt.xlabel('sales')
plt.ylabel('revenue')
plt.plot(x,y)
plt.show()
```

```python
x=[1,3,4]
y=[2,3,4]
z=[2,4,5]
plt.title('sales information-ABC')
plt.legend(['this is xy','this is xz'])
plt.xlabel('sales')
plt.ylabel('revenue')
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
x=np.arange(0,10,0.2)
```

```python
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,x,z)
plt.legend(['sine','cosine'])
plt.title('sine and cosine graph')
plt.show()
```

```python
y=np.tan(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Tangent')
plt.show()
```
